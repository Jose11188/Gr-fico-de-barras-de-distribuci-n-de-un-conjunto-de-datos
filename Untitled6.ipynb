```python
# b.	Implementar un sistema con dos pulsadores: uno para avanzar y otro para retroceder. Mostrar cinco mensajes diferentes en un LCD de manera interactiva, de modo que al presionar el bot√≥n de avance se muestre el mensaje siguiente y al presionar el bot√≥n de retroceso se muestre el mensaje anterior.

from IPython.display import clear_output
import time

def lcd_simulado_pulsadores():
    mensajes = [
        "Mensaje 1: Bienvenido",
        "Mensaje 2: Sistema Activo",
        "Mensaje 3: Verificando Datos",
        "Mensaje 4: Proceso Exitoso",
        "Mensaje 5: Fin del Programa"
    ]

    indice = 0  # √≠ndice actual del mensaje

    try:
        while True:
            # Mostrar LCD simulado
            clear_output(wait=True)
            print("+--------------------------+")
            print("|        LCD SIMULADO      |")
            print(f"| {mensajes[indice]:<24} |")
            print("+--------------------------+")
            print("\nPresiona:")
            print("  ‚û°Ô∏è  'a' para avanzar")
            print("  ‚¨ÖÔ∏è  'r' para retroceder")
            print("  ‚ùå  's' para salir")

            # Entrada del usuario
            opcion = input("Ingresa tu opci√≥n: ").strip().lower()

            if opcion == 'a':
                indice = (indice + 1) % len(mensajes)
            elif opcion == 'r':
                indice = (indice - 1) % len(mensajes)
            elif opcion == 's':
                clear_output(wait=True)
                print("‚úÖ Sistema finalizado correctamente.")
                break
            else:
                print("‚ö†Ô∏è Opci√≥n no v√°lida. Usa 'a', 'r' o 's'.")
                time.sleep(1)

    except KeyboardInterrupt:
        # Si detienes manualmente la ejecuci√≥n, no mostrar√° error
        clear_output(wait=True)
        print("‚úÖ Programa detenido por el usuario sin errores.")

# Ejecutar el programa
lcd_simulado_pulsadores()

```

```python
# c.	Cuando se presione el pulsador B1, activar de forma alternada durante 3 segundos dos LEDs, LED1 y LED2. Es decir, al presionar B1, encender LED1 durante 3 segundos. Al presionar B1 nuevamente, encender LED2 durante 3 segundos. El siguiente turno ser√° para LED1 nuevamente, y as√≠ sucesivamente.

import time

# C√≥digos de color ANSI
RESET = "\033[0m"
VERDE = "\033[92m"   # Verde para ambos LEDs

# Estado inicial (0 = LED1, 1 = LED2)
estado_led = 0

print("üí° Simulaci√≥n del pulsador B1 y LEDs verdes")
print("Opciones:")
print("  b ‚Üí Encender siguiente LED (alternado)")
print("  a ‚Üí Encender ambos LEDs al mismo tiempo")
print("  salir ‚Üí Finalizar el programa\n")

while True:
    pulsador = input("Presiona 'b', 'a' o escribe 'salir': ").strip().lower()

    if pulsador == "salir":
        print("üîö Simulaci√≥n finalizada correctamente.")
        break

    elif pulsador == "b":
        if estado_led == 0:
            print(f"{VERDE}LED1 encendido (üü¢){RESET}")
            time.sleep(3)
            print("LED1 apagado\n")
            estado_led = 1
        elif estado_led == 1:
            print(f"{VERDE}LED2 encendido (üü¢){RESET}")
            time.sleep(3)
            print("LED2 apagado\n")
            estado_led = 0

    elif pulsador == "a":
        print(f"{VERDE}LED1 y LED2 encendidos simult√°neamente (üü¢üü¢){RESET}")
        time.sleep(3)
        print("Ambos LEDs apagados\n")

    else:
        print("Entrada no v√°lida. Usa 'b', 'a' o 'salir'.\n")

```
